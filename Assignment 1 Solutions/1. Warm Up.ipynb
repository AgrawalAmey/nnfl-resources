### Problem 1.1.1

Complete the following function to return true if the given string *s* is a palindrome. Use string methods.

````python
def is_palindrome_str(s):
    # Your code goes here
    # There is no string method to 'reverse' a string,
    # but you can use a trick instead.
    if s == ''.join(reversed(s)):
        return True 
    return False
````

### Problem 1.1.2

Complete the following function to return true if the given string *s* is a palindrome. Use slicing.

````python
def is_palindrome_slicing(s):
    # Your code goes here
    if s == s[::-1]:
        return True
    return False
````

### Problem 1.1.3

Complete the following function so as to return True if the input string *s* contains substring *sub*. Make the function case insensitive and do not use loops. You can find a list of all string methods in the [documentation](https://docs.python.org/3/library/stdtypes.html#string-methods).

````python
def find_substring(s, sub):
    # Your code goes here
    if sub.lower() in s.lower():
        return True
    return False
````

Write a function which takes a 2D list as input and returns a flat list. Use only list comprehensions.

*Example*:
Input: [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
Output: [0, 1, 2, 3, 4, 5, 6, 7, 8]

````python
def flatten(matrix):
    # Your code goes here
    return [j for i in matrix for j in i]
````

### Problem 1.1.5

Write a function which takes in a 2D list as input and returns a dialted matrix i.e matrix such that only elements which have both x and y indexs even are preserved. Solve using only list comprehsions.


**Hint:** Look into enumeration in list comprehensions.

````python
def dilate(matrix):
    # Your code goes here
    return [[v for i, v in enumerate(w) if i % 2 == 0] for j, w in enumerate(matrix) if j % 2 == 0]
````

### Problem 1.1.6

Write a function which takes a string as input and returns a dictonary such that the letters from the string are the keys and their counts are the values. 

*Example*:

Input: 'helLo'
Output: {'h': 1, 'e': 1, 'l': 2:, 'o': 1}

````python
def get_letter_count(s):
    # Your code goes here
    s = s.lower()
    counts = {}
    for char in s:
        if char in counts:
            counts[char] += 1
        else:
            counts[char] = 1
    return counts
````
